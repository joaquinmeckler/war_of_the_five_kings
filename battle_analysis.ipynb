```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from pylab import rcParams

pd.set_option('display.max_columns', 50)
%matplotlib inline

rcParams['figure.figsize'] = 12, 7
```

```python
data = 'https://raw.githubusercontent.com/chrisalbon/war_of_the_five_kings_dataset/master/5kings_battles_v1.csv'
df = pd.read_csv(data)
```

```python
df.head()
```

```python
df['attacker_king'].value_counts()
```

```python
king = [item for item in df['attacker_king']]
battle_king = []

for item in king:
    if item == 'Joffrey/Tommen Baratheon':
        battle_king.append('gold')
    elif item == 'Robb Stark':
        battle_king.append('grey')
    elif item == 'Balon/Euron Greyjoy':
        battle_king.append('black')
    elif item == 'Stannis Baratheon':
        battle_king.append('red')
    else:
        battle_king.append('white')
```

```python
plt.style.use('bmh')

axis_parameters = [0, 40, -1500, 105000]
plt.axis(axis_parameters)

plt.title('Game of Thrones Battles: Comparative Forces')
plt.xlabel('Battle Number')
plt.ylabel('Number of Combatants')

plt.scatter(df['battle_number'], (df['attacker_size'] + df['defender_size']),
            c = battle_king,
            s = 150)
plt.plot([27, 27],[-1500, 105000], color = 'blue')


plt.annotate(
        df['name'][27], 
        xy = (df['battle_number'][27], (df['attacker_size'] + df['defender_size'])[27]), xytext = (10, -15),
        textcoords = 'offset points', ha = 'left', va = 'top',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'grey', alpha = 0.5),
        arrowprops = dict(facecolor='black', shrink=.1, width = 2, frac = .5))
plt.annotate(
        df['name'][3], 
        xy = (df['battle_number'][3], (df['attacker_size'] + df['defender_size'])[3]), xytext = (5, 35),
        textcoords = 'offset points', ha = 'left', va = 'top',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'grey', alpha = 0.5),
        arrowprops = dict(facecolor='black', shrink=.1, width = 2, frac = .5))
plt.annotate(
        df['name'][25], 
        xy = (df['battle_number'][25], (df['attacker_size'] + df['defender_size'])[25]), xytext = (-100, 25),
        textcoords = 'offset points', ha = 'left', va = 'top',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'grey', alpha = 0.5),
        arrowprops = dict(facecolor='black', shrink=.1, width = 2, frac = .5))
plt.annotate(
        'Winter Begins', 
        xy = (27, 60000), xytext = (20, 0),
        textcoords = 'offset points', ha = 'left', va = 'top',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'grey', alpha = 0.5),
        arrowprops = dict(facecolor='black', shrink=.1, width = 2, frac = .5))

Baratheon = matplotlib.patches.Patch(color='gold', label='Joffrey/Tommen Baratheon')
Stark = matplotlib.patches.Patch(color = 'grey', label = 'Robb Stark')
Stannis = matplotlib.patches.Patch(color = 'red', label = 'Stannis Baratheon')
plt.legend(title = 'Attacking Faction King',handles=[Baratheon, Stark, Stannis], loc = 'upper left')
```
